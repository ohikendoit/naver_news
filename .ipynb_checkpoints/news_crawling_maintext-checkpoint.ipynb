```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
df_news = pd.read_csv('data/Macquarie_19900101-20220420.csv')
```

```python
df_news
```

```python
df_news = df_news.iloc[:20]
```

```python
pd.set_option('display.max_colwidth', None)
df_news['URL']
```

```python
# Loop through URLs, scrap information, identify the main body
# 뉴스기사전문
for i in df_news['URL']:
    html = requests.get(i).text
    soup = BeautifulSoup(html, "html.parser", from_encoding='utf-8')
    print(soup)
```

```python
df_news['html_requested'] = df_news.URL.apply(lambda url: requests.get(url).status_code)
df_news
```
